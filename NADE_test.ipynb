# Plotting method

```python
# Plotting method once we have both raw data and pseudo data generated by NADE
def plot(raw_data, pseudo_data, feature_list, feat = None, nbins = 30, tail = 8, tailbins = 1):
    data_bins_list = {}
    nade_bins_list = {}
    for feature in range(raw_data.shape[1]):
        if not feat or feature == feat: 
            print "feature", feature_list[feature]
            if feature_list[feature] == 'jets_eta:0' or feature_list[feature] == 'jets_eta:1':
                spam = (raw_data[:, feature].min(), raw_data[:,feature].max())
                (data_n, data_bins, data_patches) = plt.hist( raw_data[:,feature], nbins, histtype='stepfilled', label='input',alpha=0.5, color='blue',normed = True)
                (nade_n, nade_bins, nade_patches) = plt.hist(pseudo_data[feature,:], nbins, range = spam, histtype = 'stepfilled', alpha = 0.5, label = 'part_nade', color = 'red', normed = True)
            else: 
                bins = np.zeros(nbins)
                bins[0] = raw_data[:, feature].min()
                bins[nbins-1] = raw_data[:,feature].max()
                spam = bins[nbins-1] - bins[0] 
                for n in range(tailbins): 
                    bins[nbins-2-n] = bins[nbins-1] - spam/tail
                gap = spam * (float(tail-tailbins)/float(tail))/(nbins-tailbins)
                for i in range(1, nbins-tailbins): 
                    bins[i] = bins[i-1] + gap
                (data_n, data_bins, data_patches) = plt.hist( raw_data[:,feature], bins, histtype='stepfilled', label='input',alpha=0.5, color='blue',normed = True)
                # plt.hist( pseudo_data[feature,:], 200, histtype='stepfilled', alpha=0.5, label='full_nade', color='red',normed=True)
                (nade_n, nade_bins, nade_patches) = plt.hist(pseudo_data[feature,:], bins, histtype = 'stepfilled', alpha = 0.5, label = 'part_nade', color = 'red', normed = True)
            data_bins_list[(feature)] = (data_n, data_bins) 
            nade_bins_list[(feature)] = (nade_n, nade_bins)
            # plt.yscale('log')
            plt.legend()
            plt.show()
        else: continue
    return (data_bins_list, nade_bins_list)

# Create nade data from hdf5 files
def create(src):
    if not os.path.isfile(src):
        return (None,0,0)
    
    f=h5py.File(src)
    p5=f.get("final_model").get("parameters")
    means=[]
    rmss=[]
    if f.get('normalisation/mean'):
        means = f.get('normalisation/mean').value
        print means
    if f.get('normalisation/std'):
        rmss = f.get('normalisation/std').value
        print rmss
    p={}
    for k in p5.keys():
        p[k] = p5.get(k).value

    clsn = p.get("__class__")
    cls = getattr(NADE,clsn)
    nade = cls.create_from_params( p )
    return (nade,means,rmss)

# create pseudo data from a file
def nade_pseudo(dout, nsampling = 1000 ):
    nade,means,rmss = create(dout)
    if not nade:
        return (None,nade, means, rmss)
    
    pseudo_data = nade.sample( nsampling ) 
    if means!=[] and rmss!=[]:
        pseudo_data = pseudo_data * rmss.reshape((len(rmss),1)) + means.reshape((len(means),1))
    else:
        print "no normalization" 
    return (pseudo_data, nade, means, rmss)

# A plotting method that takes a nade and do a compare plot between raw data and generated data
def compare(nade, train = None, nsampling = 1000):
    din = nade.args['datasetspath']+'/'+nade.args['dataset']
    print din
    
    if train:
        dout = '%s/%s/NADE_%d.hdf5'%( nade.args['resultspath'], nade.args['label'], train)
    else:
        dout = filter(None, os.popen('ls %s/%s/NADE*.hdf5'%( nade.args['resultspath'], nade.args['label'])).read().split('\n'))[-1]
    print dout
    
    return io_compare(din, dout, nsampling)

# A method that extracts raw_data and pseudo_data
def io_compare(din, dout, nsampling = 1000): 
    raw_data = pickle.load(open(din)) 
    (pseudo_data, nade2, means, rmss) = nade_pseudo(dout, nsampling)
    return (raw_data, pseudo_data, nade2, means, rmss)
```

# Remove Zeros

```python
## A function that only keeps columns that don't have a single zero
def remove_zeros( dataset) :
    goods=set(range(dataset.shape[1]))
    bads=set()
    ngood=0
    for s in dataset:
        good=True
        bad_f=None
        for (i,feature) in enumerate(s):
            if feature == 0 : 
                good=False
                bads.add(i)
        if good:
            #print s
            ngood+=1
            pass
        else:
            #print bad_f,"not good"
            pass
            #bads.add(bad_f)
    remain=set(list(goods-bads))
    if not ngood:
        print "not a single event has all feature non zero"
    else:
        print ngood,"sample with no zero"
    #print "good : features with no zeros",goods
    #print "bad : features with zeros",bads
    #print "keeping :",remain
    g_dataset = np.array(dataset[...,list(remain)],order='C')
    return (g_dataset, remain)

def io_remove_zeros( dsn ):
    dataset = pickle.load(open('../Data/%s'%dsn))
    (g_dataset, remain) = remove_zeros( dataset)
    new_filename = dsn.replace('.pkl','_nz.pkl')
    pickle.dump(g_dataset, open('../Data/%s'%new_filename, 'w'))
    return (new_filename, remain)
```
