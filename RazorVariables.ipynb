```python
import os
import numpy as np
import h5py
from scipy.stats import poisson
import time
from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm, NoNorm
```

```python
# A method that returns a function of mu, un-normalised
def negloglikelihood(bkg, sig, bkg_weights, sig_weight):
    '''
    bkg: number of each type of background events in each cell
    sig: number of signal events in each cell 
    bkg weights: the ratios we need to use to adjust them to realistic ratios 
    sig weight: the fraction we need to multiply the signal by to make it a function of mu
    '''
    
    # Add weights to the arrays 
    assert bkg.shape[-1] == len(bkg_weights)
    
    b = np.sum(bkg * np.array(bkg_weights), axis = -1)
    s = np.sum(sig * np.array(sig_weight), axis = -1)
    print 'background:', np.sum(b) 
    print 'signal:', np.sum(s) 
    
    assert b.shape == s.shape 
    print 'number of zero cells in bkg:', np.sum(b == 0) 
    print 'zero cells in sig:', s[b==0]
    
    b[b==0] = np.max(bkg_weights)
    print 'zero cells are replaced with', np.max(bkg_weights)
    
    print 'signal cells in bkg:', b[s!= 0] 
    print 'signal cells in sig:', s[s!= 0]
    
    return (lambda mu: 2 * np.sum((b + mu * s) - b * np.log(b + mu * s))) 

def negloglikelihood_plot(mu_range, fun, n = 100):
    ''' 
    a function that takes the outcome of the likelihood function and plot the likelihood within a certain range of mu
    mu_range: a tuple
    fun: a function
    ''' 
    mu_i, mu_f = mu_range
    mu = np.linspace(mu_i, mu_f, n)
    y = map(fun, mu) - fun(0)
    ref = np.full(y.shape, 2 * 1.29)
    
    plt.plot(mu, y)
    plt.plot(mu, ref)
    plt.title('negative likelihood plot')
    plt.xlabel('mu')
    plt.ylabel('- 2 * log(likelihood)')
    plt.show() 
    
    return y

def razorbins(samples, types, plot = False):
    '''
    x axis: MR 
    y axis: Rsq
    MR: column - 2
    Rsq: column - 3
    '''
    
    xedges = [np.min(samples[:, -2]), 400, 500, 600, 700, 900, 1300, np.max(samples[:, -2])]
    yedges = [np.min(samples[:, -3]), 0.1, 0.15, 0.20, 0.36, np.max(samples[:, -3])]

    hist = np.zeros((len(xedges)-1, len(yedges)-1, types))
    
    for t in range(types):
        MR = samples[:, -2][samples[:, -1] == t]
        Rsq = samples[:, -3][samples[:, -1] == t]
        hist[..., t], _, _, _ = plt.hist2d(MR, Rsq, bins = [xedges, yedges])
        plt.colorbar() 
        plt.show()
    return hist     

# View the data excluding the zeros
def view_labeled(data, bin_number, labels, feature_list):
    colors= ['red','blue','green','orange','purple','yellow','brown']
    for feature in range(len(feature_list)):
        cs=[]
        ls=[]
        ms=[]
        plt.figure(figsize=(10, 5))
    
        for t in range(len(labels)):
            matching = [d[feature] for d in data if (d[-1]==t)]
            plt.subplot(1,2,1)
            plt.hist( matching, bin_number, color = colors[t] , histtype='stepfilled', alpha=0.5, label=labels[t])
            plt.xlabel(feature_list[feature])
            cs.append( colors[t] )
            ls.append( labels[t] )
            ms.append( matching )
        print "feature",feature_list[feature]
        plt.legend()
        plt.subplot(1,2,2)
        ms.reverse()
        cs.reverse()
        ls.reverse()
        plt.hist( ms , bin_number, color=cs, label=ls, histtype='stepfilled', alpha=0.5, stacked=True)
        plt.xlabel(feature_list[feature])
        plt.legend()
        plt.show()
```
