	Properties of the map: 
		- Variables
			- features 
			- dim: dimensionality of the map 
			- units: number of units in the map 
		- theano symbolic shared variables
			- shape: shape of the map
			- grid: coordinates of all the cells in the map
			- codebook: weights of all the cells

	Training parameters: 
		- Variables
			- epochs
			- lrate_i, lrate_f
			- sigma_i, sigma_f 
			- threshold
		- theano symbolic shared variables
			- sigma: contains the current sigma 
			- lrate: contains the current lrate 
			- sigma_factor: the factor by which sigma gets updated at each epoch
			- lrate_factor: the factor by which lrate gets updates at each epoch

	Properties of the datasets:
		- Variables
			- means: means of the datasets
			- stds: standard deviation of the datasets
			- fires: index of the cell that fires (produced in training)
			- test: hits in each cell by different types of data (produced in testing)

	Functions: 
		- __init__(shape, features, filename = None)
			- Initialise the map 
		- set_params(epochs = 5, sigma = (6, 0.001), lrate = (0.2, 0.001), threshold = 4)
			- Set training parameters 
		- train_theano(data): 
			- Rescale the data and train the map
		- test_theano(data, types):
			- Test with data
		- save_map(filename)
			- Save the entire content of the map
		- save_results(filename) 
			- Save the results from testing

```python
# A method that selects and rescale 
def select(dataset, samples, weights, n_features): # dataset is a set of np arrays 
    types = len(samples)
    n_samples = np.sum(samples)
    data = np.zeros((n_samples, n_features + 2))
    
    first = 0
    for i in range(types): 
        if samples[i] != 0:
            high = dataset[i].shape[0]
            ordering = np.random.randint(0, high = high, size = samples[i]) 
            for (k, j) in enumerate(ordering):
                data[first + k , 0:n_features] = dataset[i][j] 
            data[first:first+samples[i], -2] = i
            data[first:first+samples[i], -1] = weights[i]
            first += samples[i]
    
    # shuffle the data
    np.random.shuffle(data)
    
    return data

def cut(data, veto, req, red, ptcut = 0, sumcut = 0):
    '''
    an method that perform cuts in a dataset 
    veto: reject event if feature is nonzero 
    req: reject event if feature is zero 
    red: redundant information
    ptcut: cut in metPt 
    sumcut: cut in sumMET
    '''
    samples = data.shape[0] #number of datapoints 
    features = data.shape[1] #number of features 
    total_indices = np.arange(features)
    remaining_indices = [i for i in total_indices if not i in veto and not i in red]
    
    cut_data = np.zeros((samples, len(remaining_indices)))
    count = 0 
    for i in range(samples):
        sample = data[i]
        if np.sum(sample[veto] != 0) == 0 and np.sum(sample[req] == 0) == 0: 
            if sample[-2] > ptcut and sample[-1] > sumcut:
                cut_data[count] = sample[remaining_indices]
                count += 1
    
    return cut_data, count

# A method that plots histograms of all the features 
def show_data(feature_list, yscale, *arg):
    n_args = len(arg)
    print "comparing", n_args, "datasets" 
    for (i, feat) in enumerate(feature_list):
        print feat
        plt.yscale(yscale)
        for data in arg:
            plt.hist(data[:, i], label = feat,  bins = 100, alpha = 0.4, normed = True, histtype = 'stepfilled')
        plt.show()     

# A method that compares testing and training
def compare_toPoisson(tst, trn, sigcol, withsig = False):
    trn_bkg = np.copy(trn)
    trn_bkg[..., sigcol] = 0
    tst_bkg = np.copy(tst)
    tst_bkg[..., sigcol] = 0 
    
    factor = np.sum(tst_bkg)/float(np.sum(trn_bkg))
    print 'factor:', factor
    total_tst = np.ravel(np.sum(tst_bkg, axis = -1)/factor)
    total_trn = np.ravel(np.sum(trn_bkg, axis = -1))
    
    
    print 'zero cells in tst'
    print total_tst[total_tst == 0]
    print 'zero cells in trn_bkg'
    print total_trn[total_tst == 0]
    total_tst[total_tst == 0] = 1/factor #replace the zero cell
    
      
    comp = poisson.rvs(total_tst) 


    pmf = poisson.pmf(total_trn, total_tst)
    pmf_comp = poisson.pmf(comp, total_tst)

    cdf = poisson.cdf(total_trn, total_tst)
    cdf_comp = poisson.cdf(comp, total_tst)
    
    sig = (total_trn - total_tst)/np.sqrt(total_tst)
    sig_comp = (comp - total_tst)/np.sqrt(total_tst)
    
    if withsig:
        total_sig = np.ravel(np.sum(trn, axis = -1))
        sigcells = np.ravel_multi_index(np.where(trn[..., sigcol]!= 0), (10, 10, 10))
        print 'signal cells in training', total_sig[sigcells]
        print 'signal cells in testing', total_tst[sigcells]

        sig_sig = ((total_sig - total_tst)/np.sqrt(total_tst))[sigcells]
        cdf_sig = poisson.cdf(total_sig[sigcells], total_tst[sigcells])
        pmf_sig = poisson.pmf(total_sig[sigcells], total_tst[sigcells])
        

    bins, edges, patches = plt.hist(cdf, bins = 100, alpha = 0.6, color = 'orange', normed = False, histtype = 'stepfilled')
    plt.hist(cdf_comp, bins = edges, alpha = 0.4, color = 'blue', normed = False, histtype = 'stepfilled')
    if withsig:
        plt.hist(cdf_sig, bins = edges, alpha = 0.6, color = 'red', normed = False)
    plt.title('cumulative distributive function')
    plt.ylabel('number of cells')
    plt.xlabel('cdf')
    plt.ylim([1, 200])
    plt.yscale('log')
    plt.show()
    
    bins, edges, patches = plt.hist(pmf, bins = 100, alpha = 0.5, color = 'orange', histtype = 'stepfilled')
    plt.hist(pmf_comp, bins = edges, alpha = 0.4, color = 'blue', histtype = 'stepfilled')
    if withsig:
        plt.hist(pmf_sig, bins = 100, alpha = 0.7, color = 'red', histtype = 'stepfilled')
    plt.title('probability mass function')
    plt.xlabel('probability mass function')
    plt.ylabel('number of cells')
    plt.yscale('log')
    plt.ylim([0.8, 1000])
    plt.show()

    bins, edges, patches = plt.hist(np.log(pmf), bins = 100, alpha = 0.5, color = 'orange', histtype = 'stepfilled')
    plt.hist(np.log(pmf_comp), bins = edges, alpha = 0.4, color = 'blue', histtype = 'stepfilled')
    if withsig:
        plt.hist(np.log(pmf_sig), bins = 100, alpha = 0.8, color = 'red', histtype = 'stepfilled')
    plt.title('log probability mass function')
    plt.xlabel('log probability mass function')
    plt.ylabel('number of cells')
    plt.yscale('log')
    plt.ylim([0.7, 1000])
    plt.show()
    
    bins, edges, patches = plt.hist(np.ravel(sig), bins = 100, alpha = 0.4, color = 'orange', histtype = 'stepfilled')
    plt.hist(sig_comp, bins = edges, alpha = 0.4, color = 'blue', histtype = 'stepfilled')
    if withsig:
        plt.hist(sig_sig, bins = 100, alpha = 0.8, color = 'red', histtype = 'stepfilled')
    plt.title('deviation = (trn - tst)/sqrt(tst)')
    plt.xlabel('deviation')
    plt.ylabel('number of cells')
    plt.yscale('log')
    plt.ylim([0.8, 100])
    plt.show()
        
# A method that returns a function of mu, un-normalised
def negloglikelihood(bkg, sig, bkg_weights, sig_weight):
    '''
    bkg: number of each type of background events in each cell
    sig: number of signal events in each cell 
    bkg weights: the ratios we need to use to adjust them to realistic ratios 
    sig weight: the fraction we need to multiply the signal by to make it a function of mu
    '''
    
    # Add weights to the arrays 
    assert bkg.shape[-1] == len(bkg_weights)
    
    b = np.sum(bkg * np.array(bkg_weights), axis = -1)
    s = np.sum(sig * np.array(sig_weight), axis = -1)
    print 'background:', np.sum(b) 
    print 'signal:', np.sum(s) 
    
    assert b.shape == s.shape 
    print 'number of zero cells:', np.sum(b == 0) 
    print 'zero cells in sig:', s[b==0]
    b[b==0] = np.max(bkg_weights)
    print 'zero cells are replaced with', np.max(bkg_weights)
    
    print 'signal cells in bkg:', b[s!= 0] 
    print 'signal cells in sig:', s[s!= 0]
    
    return (lambda mu: 2 * np.sum((b + mu * s) - b * np.log(b + mu * s))) 

def negloglikelihood_plot(mu_range, fun, n = 100):
    ''' 
    a function that takes the outcome of the likelihood function and plot the likelihood within a certain range of mu
    mu_range: a tuple
    fun: a function
    ''' 
    mu_i, mu_f = mu_range
    mu = np.linspace(mu_i, mu_f, n)
    y = map(fun, mu) - fun(0)
    ref = np.full(y.shape, 2 * 1.29)
    
    plt.plot(mu, y)
    plt.plot(mu, ref)
    plt.title('negative likelihood plot')
    plt.xlabel('mu')
    plt.ylabel('- 2 * log(likelihood)')
    plt.show() 
    
    return y

def hist3D(bins, grid):
    types = bins.shape[-1]
    x, y, z = grid
    for t in range(types):     
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.scatter(x, y, z, c = bins[..., t], norm = LogNorm(), lw = 0)
        plt.show()
```
