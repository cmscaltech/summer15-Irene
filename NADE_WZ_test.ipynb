```python
def replace_zeros(dataset): 
    samples = dataset.shape[0] 
    features = dataset.shape[1] 
    maxs = dataset.max(0)
    mmax = maxs.max()
    dataset[dataset == 0] = mmax
    mins = dataset.min(0)
    stds = (maxs - mins)/6
    means = 2 * mins - maxs
    for i in range(features): 
        for j in range(samples):
            if dataset[j, i] == mmax:
                dataset[j, i] = np.random.normal(means[i], stds[i])
    return (mins, dataset)
```

```python
# Split the data into training set and testing set according to ratio
def split(dataset, ratio = 0.5): 
    dim = len(dataset)
    trndata = dataset[0:ratio * dim, :] 
    tstdata = dataset[ratio * dim + 1: dim -1, :] 
    return trndata, tstdata
```

```python
# Plotting method once we have both raw data and pseudo data generated by NADE
def plot(raw_data, pseudo_data, feature_list, list_of_mins, feat = None, show = False, nbins = 30, tail = 8, tailbins = 1):
    data_bins_list = {}
    nade_bins_list = {}
    
    #Get the scale factor between raw data and pseudo data 
    norm_factor = float(pseudo_data.shape[1])/raw_data.shape[0] 
    
    for feature in range(raw_data.shape[1]):
        if not feat or feature == feat: 
            #Generate the bins for the histogram
            bins = np.zeros(nbins)
            bins[0] = list_of_mins[feature]
            bins[nbins-1] = pseudo_data[feature, :].max() 
            spam = bins[nbins-1] - bins[0] 
            if feature_list[feature] == 'jets_eta:0' or feature_list[feature] == 'jets_eta:1' or feature_list[feature] == 'jets_eta:2': #special case because fat tail
                temp_tail = nbins
                temp_tailbins = 0
            else: 
                temp_tail = tail
                temp_tailbins = tailbins 

            for n in range(temp_tailbins): 
                bins[nbins-2-n] = bins[nbins-1] - spam/temp_tail
            gap = spam * (float(temp_tail-temp_tailbins)/float(temp_tail))/(nbins-temp_tailbins)
            for i in range(1, nbins-temp_tailbins): 
                bins[i] = bins[i-1] + gap
            # Obtain the hist and rescale raw data
            data_hist, data_bins = np.histogram(raw_data[:, feature], bins) 
            nade_hist, nade_bins = np.histogram(pseudo_data[feature, :], bins)
            data_hist *= norm_factor
            data_bins_list[(feature)] = (data_hist, data_bins) 
            nade_bins_list[(feature)] = (nade_hist, nade_bins)
            
            if show: 
                print "feature", feature_list[feature]
                # PLOT
                left = bins[0:nbins-1] 
                right = bins[1:nbins]
                width = right - left
                data_height = data_hist/width
                nade_height = nade_hist/width

                plt.bar(left, data_height, width = width, label = 'test data', alpha = 0.4, color = 'LightPink', edgecolor = 'none')
                plt.bar(left, nade_height, width = width, label = 'nade', alpha = 0.4, color = 'LightBlue', edgecolor = 'none')
                plt.yscale('log')
                plt.legend()
                plt.show()
        else: continue
    return (data_bins_list, nade_bins_list)
```

```python
# Create nade data from hdf5 files
def create(src):
    if not os.path.isfile(src):
        print 'not a file'
        return (None,0,0)
    
    f=h5py.File(src)
    p5=f.get("final_model").get("parameters")
    means=[]
    rmss=[]
    if f.get('normalisation/mean'):
        means = f.get('normalisation/mean').value
        print means
    if f.get('normalisation/std'):
        rmss = f.get('normalisation/std').value
        print rmss
    p={}
    for k in p5.keys():
        p[k] = p5.get(k).value

    clsn = p.get("__class__")
    cls = getattr(NADE,clsn)
    nade = cls.create_from_params( p )
    return (nade,means,rmss)

# create pseudo data from a file
def nade_pseudo(dout, nsampling = 1000 ):
    nade,means,rmss = create(dout)
    if not nade:
        print "Not NADE"
        print means, rmss
        return (None,nade, means, rmss)
    
    pseudo_data = nade.sample( nsampling ) 
    if means!=[] and rmss!=[]:
        pseudo_data = pseudo_data * rmss.reshape((len(rmss),1)) + means.reshape((len(means),1))
        print pseudo_data[0][0]
    else:
        print "no normalization" 
    return (pseudo_data, nade, means, rmss)

# A plotting method that takes a nade and do a compare plot between raw data and generated data
def extract_pseudo(nade, train = None, nsampling = 1000):
    if train:
        dout = '%s/%s/NADE_%d.hdf5'%( nade.args['resultspath'], nade.args['label'], train)
    else:
        dout = filter(None, os.popen('ls %s/%s/NADE*.hdf5'%( nade.args['resultspath'], nade.args['label'])).read().split('\n'))[-1]
    print dout 
    return nade_pseudo(dout, nsampling)
```

```python
def mean(array):
    s = 0
    l = 0
    for x in array: 
        if not np.isinf(x) and not np.isnan(x): 
            s += x
            l += 1
    if l == 0: 
        return float('nan')
    else: 
        return s/float(l)

def std(array):
    s = 0
    l = 0 
    for x in array:
        if not np.isinf(x) and not np.isnan(x):
            s += x ** 2
            l += 1
    if l == 0: 
        return float('nan')
    else: 
        return np.sqrt((s/float(l) - mean(array)**2))
```

```python
def weighted(dataset): 
    weights = (1/dataset[:, :, 1])**2
    w_means, weight_sums = np.average(dataset[:, :, 0], axis = 1, weights = weights, returned = True)
    return w_means, np.sqrt(1/weight_sums)
```
