```python
def cut(data, veto, req, red, ptcut = 0, sumcut = 0):
    '''
    an method that cuts events out if the elements at indices are not all zeros 
    '''
    samples = data.shape[0] #number of datapoints 
    features = data.shape[1] #number of features 
    total_indices = np.arange(features)
    remaining_indices = [i for i in total_indices if not i in veto and not i in red]
    
    cut_data = np.zeros((samples, len(remaining_indices)))
    count = 0 
    for i in range(samples):
        sample = data[i]
        if np.sum(sample[veto] != 0) == 0 and np.sum(sample[req] == 0) == 0: 
            if sample[-2] > ptcut and sample[-1] > sumcut:
                cut_data[count] = sample[remaining_indices]
                count += 1
    
    return cut_data, count

# a function tha takes a list of ranges and cut the data within the ranges 
def cut_range(data, ranges):
    assert data.shape[1] == len(ranges)
    boo = np.zeros(data.shape[0], dtype = bool)
    
    def in_range(x, (a, b)):
        if a <= x <= b:
            return True
        else:
            return False
    
    for i in range(data.shape[0]):
        if np.sum([not in_range(x, y) for (x, y) in zip(data[i], ranges)]) == 0:
            boo[i] = 1
    
    return data[boo], data[np.invert(boo)]
    

# a function that takes a list of functions and apply them to the data 
def transform(data, funcs):
    assert data.shape[1] == len(funcs)
    data_copy = np.zeros(data.shape)
    for i in range(data.shape[1]):
        data_copy[:, i] = funcs[i] (data[:, i])
    return data_copy

# A method that splits the data into training and testing, and save the arrays to files 
def split(data, label, filename, ratio = 0.5, io = True, show = False): 
    
    # split into right ratios 
    np.random.shuffle(data)
    dim = len(data)
    trndata = data[0:ratio * dim, :] 
    cvdata = data[ratio * dim: ratio * dim * 2, :]
    tstdata = data[ratio * dim * 2: dim-1, :]
    
    io_training = None 
    # Dump into hdf5 files
    if io:
        print label
        io_training = '../Data/%s_%s.hdf5'%(filename, label) 
        print io_training
        f = h5py.File(io_training, 'a')
        print f.keys()
        if "train" in f.keys():
            print 'train exists'
            grp = f['train']
            del grp['data']
            dset = grp.create_dataset('data', data= trndata)
            del f['test']['data']
            del f['validation']['data']
            f.create_dataset('test/data', data = tstdata)
            f.create_dataset('validation/data', data = cvdata)
        else:
            print 'creating train'
            grp = f.create_group('train')
            grp.create_dataset('data', data = trndata)
            f.create_group('validation').create_dataset('data', data = cvdata)
            f.create_group('test').create_dataset('data', data = tstdata) # don't waste too much data on testing 
        f.close() 
    
    # Check if everything makes sense 
    if show: 
        for i in range(data.shape[-1]):
            print 'feature:', i
            plt.hist(trndata[:, i], bins = 100, alpha = 0.4, normed = True, histtype = 'stepfilled')
            plt.hist(tstdata[:, i], bins = 100, alpha = 0.4, normed = True, histtype = 'stepfilled')
            plt.hist(cvdata[:, i], bins = 100, alpha = 0.4, normed = True, histtype = 'stepfilled')
            plt.show() 
    
    return trndata, tstdata, io_training

# A method that randomly takes a fracture of the data
def take(data, ratio): 
    np.random.shuffle(data)
    dim = len(data)
    data = data[0:ratio * dim, :] 
    return data 

# Create nade data from hdf5 files
def create(src):
    if not os.path.isfile(src):
        print 'not a file'
        return (None,0,0)
    
    f=h5py.File(src)
    p5=f.get("final_model").get("parameters")
    means=[]
    rmss=[]
    if f.get('normalisation/mean'):
        means = f.get('normalisation/mean').value
    if f.get('normalisation/std'):
        rmss = f.get('normalisation/std').value
    p={}
    for k in p5.keys():
        p[k] = p5.get(k).value

    clsn = p.get("__class__")
    cls = getattr(NADE,clsn)
    nade = cls.create_from_params( p )
    return (nade,means,rmss)

# Get a model from a oNADE object
def getModel(nade):
    directory = '%s/%s'%(nade.args['resultspath'], nade.args['label'])
    files = [os.path.join(directory,d) for d in os.listdir(directory)]
    training = max(files, key=os.path.getmtime)
    print training
    files.remove(training)
    dout = max(files, key = os.path.getmtime)
    print dout
    return (create(dout), dout)
    
# Get the pseudo data and recover the zeros 
def extract(model, nsampling = 1000): 
    nade, means, rmss = model
    pseudo = (nade.sample(nsampling))
    
    means = means.reshape(len(means), 1)
    rmss = rmss.reshape(len(rmss), 1)
    
    return pseudo * rmss + means 

# A method that plots histograms of all the features 
def show_data(feature_list, yscale, *arg):
    n_args = len(arg)
    print "comparing", n_args, "datasets" 
    for (i, feat) in enumerate(feature_list):
        print feat
        plt.yscale(yscale)
        for data in arg:
            plt.hist(data[:, i], label = feat,  bins = 100, alpha = 0.4, normed = True, histtype = 'stepfilled')
        plt.show()     

# A method that takes a model, rescale the tstdata according to it and generate pseudo data with same dim as tstdata
def compare(model, data, density = True, features = None): 
    
    # Unpack the model tuple
    nade, means, rmss = model
    means = means.reshape(len(means), 1)
    rmss = rmss.reshape(len(rmss), 1)
    
    # Rescale data
    data = ((data.T - means)/rmss)
    pseudo = nade.sample(data.shape[1])
    
    # Print out loglikelihood
    estimate = nade.estimate_loglikelihood_for_dataset(Data.Dataset(data.T))
    print 'average log likelihood:', estimate.estimation
    print 'standard deviation:', estimate.se
    
    # Plot log density
    if density:
        logdensity = nade.logdensity(data)
        print 'anomalies:', logdensity[logdensity < -50]
        hist, edges, batches = plt.hist(logdensity, label = 'data', bins = 100, alpha = 0.4, color = 'blue', histtype = 'stepfilled')
        plt.hist(nade.logdensity(pseudo), label = 'pseudo', bins = edges, alpha = 0.4, color = 'red', histtype = 'stepfilled')
        plt.yscale('log')
        plt.xlabel('log density')
        plt.ylabel('number of data points')
        plt.ylim([0.5, 1000])
        plt.legend()
        plt.show()
        
    # Plot histgrams if features 
    if features != None:
        for i in range(len(features)):
            print 'feature:', features[i]
            hist, edges, patches = plt.hist(data[i, :], bins = 100, alpha = 0.5, histtype = 'stepfilled')
            plt.hist(pseudo[i, :], bins = edges, alpha = 0.5, histtype = 'stepfilled')
            plt.show()

    return data, pseudo  

# A method that takes a model and return a loglikelihood test object 
def negloglikelihood(model, bkg, sig, bins = 10, bkg_weight = 1, sig_weight = 1):
    # Unpack the model tuple
    nade, means, rmss = model
    means = means.reshape(len(means), 1)
    rmss = rmss.reshape(len(rmss), 1)
    
    # Rescale data
    bkg = ((bkg.T - means)/rmss)
    sig = ((sig.T - means)/rmss)
    bkg_logdensity = nade.logdensity(bkg)
    sig_logdensity = nade.logdensity(sig)
    

    l = np.min(bkg_logdensity)
    r = np.max(bkg_logdensity)
    print 'smallest log density:', l 
    print 'largest log density:', r
    edges = np.arange(l, r, (r-l)/float(bins))
    edges = np.concatenate(([-100], edges[1:]))
        
    
    b, edges = np.histogram(bkg_logdensity, bins = edges)
    index = 0
    indices = np.full(b.shape, 1, dtype = bool)
    while index < len(b):
        if b[index] == 0: 
            b[index+1] += b[index] 
            indices[index] = 0
        index += 1
        
    b = b[indices]  
    edges = edges[np.concatenate(([True], indices))]
    print edges
        
    s, edges = np.histogram(sig_logdensity, bins = edges)
    b = b * bkg_weight
    s = s * sig_weight
    print b
    print s

    lefts = edges[0:-1] 
    rights = edges[1:]
    widths = rights - lefts 
    
    plt.bar(lefts, b/widths, widths, alpha = 0.4, color = 'blue', label = 'background', edgecolor = 'none', linewidth = 0)
    plt.bar(lefts, s/widths, widths, alpha = 0.4, color = 'red', label = 'signal', edgecolor = 'none', linewidth = 0)
    plt.legend()
    plt.yscale('log')
    plt.title('Log Density Histogram')
    plt.xlabel('log density')
    plt.ylabel('number of data points/width')
    plt.show() 
    
    print 'zero bins in bkg:', b[b == 0]
    print 'zero bins in sig:', s[b == 0]
    
    return (lambda mu: 2 * np.sum((b + mu * s) - b * np.log(b + mu * s)))

def negloglikelihood_plot(mu_range, fun, n = 100):
    ''' 
    a function that takes the outcome of the likelihood function and plot the likelihood within a certain range of mu
    mu_range: a tuple
    fun: a function
    ''' 
    mu_i, mu_f = mu_range
    mu = np.linspace(mu_i, mu_f, n)
    y = map(fun, mu) - fun(0)
    ref = np.full(y.shape, 2 * 1.92)
    
    plt.plot(mu, y)
    plt.plot(mu, ref)
    plt.title('negative likelihood plot')
    plt.xlabel('mu')
    plt.ylabel('- 2 * log(likelihood)')
    plt.show() 
    
    return y
```
