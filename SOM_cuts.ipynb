```python
# -----------------------------------------------------------------------------
# Self-organizing map
# Copyright (C) 2011  Nicolas P. Rougier
#
# Distributed under the terms of the BSD License.
# -----------------------------------------------------------------------------
import numpy as np
import time
from threading import Thread

def fromdistance(fn, shape, center=None, dtype=float):
    def distance(*args):
        d = 0
        for i in range(len(shape)):
            d += ((args[i]-center[i])/float(max(1,shape[i]-1)))**2
        return np.sqrt(d)/np.sqrt(len(shape))
    if center == None:
        center = np.array(list(shape))//2
    return fn(np.fromfunction(distance,shape,dtype=dtype))

def Gaussian(shape,center,sigma=0.5):
    ''' '''
    def g(x):
        return np.exp(-x**2/sigma**2)
    return fromdistance(g,shape,center)

class SOM(Thread):
    ''' Self-organizing map '''
    
    '''
    codebook: a map of the all the weights 
    dim: dimension of the map
    tedious: a boolean value indicating something.. 
    epochs: number of epochs 
    sigma: sigma_i, sigma_f 
    lr: lr_i, lr_f 
    samples: samples 
    features: number of features of the sample
    fires: the winner neuron for each sample, number of times and the class of the data point
    test: the neurons hit during test
    ''' 

    def __init__(self, shape, ted=False):
        Thread.__init__(self)
        ''' Initialize som '''
        self.codebook = np.zeros(shape)
        self.dim = len(shape)-1 
        self.tedious = ted
        #self.events = np.full(shape[0:len(shape)-1], [])
        self.reset()

    def reset(self):
        ''' Reset weights '''
        self.codebook = np.random.random(self.codebook.shape)

    def winner( self, data):
        return self.codebook[self.winner_index( data )]
        
    def winner_index(self, data):
        D = ((self.codebook-data)**2).sum(axis=-1)
        return np.unravel_index(np.argmin(D), D.shape)


        
    def start(self, samples, epochs=25000, sigma=(10, 0.001), lrate=(0.5,0.005)):
        ''' Learn samples '''
        self.sigma_i, self.sigma_f = sigma
        self.lrate_i, self.lrate_f = lrate
        self.epochs = epochs
        self.samples = samples
            
        
    def run(self):        
        if self.tedious:
            self.fires = [None for i in self.samples]
            print len(self.fires)
            
        start = time.mktime(time.localtime())
        pickAtRandom = True
        if self.epochs > len(self.samples):
            print "events will be taken sequentially from samples"
            pickAtRandom = False
            
        sample=-1
        while (sample<self.epochs) or (self.tedious and any(map(lambda e : e==None,self.fires))):
        ##for i in range(self.epochs):
            sample+=1
            if sample%(self.epochs/100)==0 and sample:
                now = time.mktime(time.localtime())
                self.eta = (now - start) / float(sample) * float(self.epochs-sample)
                #print "Epoch",sample,"eta %5.2f [min]"% self.eta

            # Adjust learning rate and neighborhood
            t = sample/float(self.epochs)
            lrate = self.lrate_i*(self.lrate_f/float(self.lrate_i))**t
            sigma = self.sigma_i*(self.sigma_f/float(self.sigma_i))**t

            # Get random sample
            if pickAtRandom:
                index = np.random.randint(0,self.samples.shape[0])
            else:
                index = sample%self.samples.shape[0]
            data = self.samples[index]

            # Get index of nearest node (minimum distance)
            D = ((self.codebook-data)**2).sum(axis=-1)
            winner = np.unravel_index(np.argmin(D), D.shape)

            if self.tedious:
                if winner == self.fires[index]:
                    #this event already fired that neuron : skip things
                    #i-=1
                    self.skipped+=1
                    continue
                else:
                    self.fires[index] = winner
            
            ## has the neuron been hit (=0 -> dead neuron)
            self.hits[winner] +=1
            
            # Generate a Gaussian centered on winner
            G = Gaussian(D.shape, winner, sigma)
            G = np.nan_to_num(G)

            # Move nodes towards sample according to Gaussian 
            delta = self.codebook-data
            for i in range(self.codebook.shape[-1]):
                self.codebook[...,i] -= lrate * G * delta[...,i]

        now = time.mktime(time.localtime())
        self.cook_rate = (now - start) / self.epochs  ## s per epochs
        self.cook_time = (now - start) ## s
   
    def control_train(self, samples, types, epochs= 5, sigma=(10, 0.001), lrate=(0.5,0.005), threshold = 4):
        
        # initialise the paramters 
        self.fires = np.zeros((samples.shape[0], self.dim + 2)) #one for number of times and one for type
        self.sigma_i, self.sigma_f = sigma
        self.lrate_i, self.lrate_f = lrate
        self.epochs = epochs
        self.types = types 
        
        features = self.codebook.shape[-1] # one for type, one for weight
        print 'number of features:', features
                
        for i in range(epochs): 
            t = i/float(epochs)
            sigma = self.sigma_i*(self.sigma_f/float(self.sigma_i))**t
            lrate = self.lrate_i*(self.lrate_f/float(self.lrate_i))**t
            
            start = time.mktime(time.localtime())
            print 'epoch:', i
            print 'learning rate:', lrate
            print 'sigma:', sigma
            
            for index in range(samples.shape[0]):
                
                # Adjust learning rate and neighborhood
                weight = samples[index, -1]
                
                data = samples[index, 0:features]
        
                # Forward pass 
                D = ((self.codebook-data)**2).sum(axis=-1)
                winner = np.unravel_index(np.argmin(D), D.shape)

                if np.array_equal(self.fires[index, 0:self.dim], winner):
                    self.fires[index, self.dim] += 1
                
                else:
                    self.fires[index, 0:self.dim] = winner
                    self.fires[index, self.dim] = 0
                    self.fires[index, self.dim + 1] = samples[index, features]
                 
                
                if self.fires[index, self.dim] < threshold: 
                    
                    # Move nodes towards sample according to Gaussian 
                    delta = self.codebook-data 
                
                    G = Gaussian(D.shape, winner, sigma)
                    G = np.nan_to_num(G)
             
                    for j in range(features):
                        delta[..., j] *=  (1 - lrate * G) ** weight
        
                    self.codebook = data + delta
                            
            
            # One epoch of training has ended 
            end = time.mktime(time.localtime())
            #print 'end time:', end 
            print 'time per epoch:', end - start
            #print 'time per sample:', (end - start)/samples.shape[0] 
            print 'number of stable samples:', np.sum(self.fires[:, self.dim] >= threshold) 
            
    def control_test(self, tst):
        start = time.mktime(time.localtime())
        self.test = np.zeros(self.codebook.shape[0:-1] + tuple([self.types]))
        features = self.codebook.shape[-1]
        
        for index in range(tst.shape[0]):
            data = tst[index, 0:features]
            weight = tst[index, -1]

            # Forward pass 
            D = ((self.codebook-data)**2).sum(axis=-1)
            winner = np.unravel_index(np.argmin(D), D.shape)

            self.test[winner + tuple([tst[index, features]])] += weight
        end = time.mktime(time.localtime())
        print 'total time:', end - start
        print 'time per sample:', (end - start)/tst.shape[0] 
        
    # A plotting method that shows 3d histgrams   
    def show_map(self, threshold = 0):
        bins = self.test
        dim = bins.shape[0]
        types = bins.shape[-1]
        x, y, z = np.indices((dim, dim, dim))
        for t in range(types):     
            fig = plt.figure()
            ax = fig.add_subplot(111, projection='3d')
            ax.scatter(x, y, z, c = bins[..., t], norm = LogNorm(), lw = 0)
            plt.show()
                              
    def best_cell(self): #show the amount of contamination in the least contaminated cell, need to add weights etc
        bkg = np.sum(self.test[..., 1:self.test.shape[-1]], -1)
        bkg[bkg == 0] = 0.01 #replace zero with 0.01, add in uncertainties later
        signif = self.test[..., 0]/(np.sum(self.test[..., 0]) * np.sqrt(bkg))
        best_cell = np.unravel_index(np.argmax(signif), self.test.shape[0:-1])
        return best_cell, self.test[best_cell][0] 
    
    def sig_cells(self): #Output a list of signal cells 
        return np.where(DataMap.test[..., 0]!= 0)
```

```python
# A method that selects and rescale 
def select(dataset, samples, weights, n_features, scaler = None): # dataset is a set of np arrays 
    '''
    dataset: a set of np_arrays
    samples: the number of samples from each type 
    weights: weights of each sample 
    n_features: number of features 
    scaler: an object that performs scaling
    '''
    types = len(samples)
    n_samples = np.sum(samples)
    data = np.zeros((n_samples, n_features + 2))
    
    first = 0
    for i in range(types): 
        if samples[i] != 0:
            high = dataset[i].shape[0]
            ordering = np.random.randint(0, high = high, size = samples[i]) 
            for (k, j) in enumerate(ordering):
                data[first + k , 0:n_features] = dataset[i][j] 
            data[first:first+samples[i], -2] = i
            data[first:first+samples[i], -1] = weights[i]
            first += samples[i]
    
    # rescales the data
    if scaler == None: 
        scaler = preprocessing.StandardScaler().fit(data[:, 0:n_features]) 
    data[:, 0:n_features] = scaler.transform(data[:, 0:n_features]) + 0.5
    
    # shuffle the data
    np.random.shuffle(data)
    
    return data, scaler  

def cut(data, veto, req, red, ptcut = 0, sumcut = 0):
    '''
    an method that perform cuts in a dataset 
    veto: reject event if feature is nonzero 
    req: reject event if feature is zero 
    red: redundant information
    ptcut: cut in metPt 
    sumcut: cut in sumMET
    '''
    samples = data.shape[0] #number of datapoints 
    features = data.shape[1] #number of features 
    total_indices = np.arange(features)
    remaining_indices = [i for i in total_indices if not i in veto and not i in red]
    
    cut_data = np.zeros((samples, len(remaining_indices)))
    count = 0 
    for i in range(samples):
        sample = data[i]
        if np.sum(sample[veto] != 0) == 0 and np.sum(sample[req] == 0) == 0: 
            if sample[-2] > ptcut and sample[-1] > sumcut:
                cut_data[count] = sample[remaining_indices]
                count += 1
    
    return cut_data, count

# a function that takes a list of functions and apply them to the data 
def transform(data, funcs):
    assert data.shape[1] == len(funcs)
    data_copy = np.zeros(data.shape)
    for i in range(data.shape[1]):
        data_copy[:, i] = funcs[i] (data[:, i])
    return data_copy

# A method that plots histograms of all the features 
def show_data(feature_list, yscale, *arg):
    n_args = len(arg)
    print "comparing", n_args, "datasets" 
    for (i, feat) in enumerate(feature_list):
        print feat
        plt.yscale(yscale)
        for data in arg:
            plt.hist(data[:, i], label = feat,  bins = 100, alpha = 0.4, normed = True, histtype = 'stepfilled')
        plt.show()     

# A method that compares testing and training
def compare_toPoisson(tst, trn, sigcol, withsig = False):
    trn_bkg = np.copy(trn)
    trn_bkg[..., sigcol] = 0
    tst_bkg = np.copy(tst)
    tst_bkg[..., sigcol] = 0 
    
    factor = np.sum(tst_bkg)/float(np.sum(trn_bkg))
    print 'factor:', factor
    total_tst = np.ravel(np.sum(tst_bkg, axis = -1)/factor)
    total_trn = np.ravel(np.sum(trn_bkg, axis = -1))
    
    
    print 'zero cells in tst'
    print total_tst[total_tst == 0]
    print 'zero cells in trn_bkg'
    print total_trn[total_tst == 0]
    total_tst[total_tst == 0] = 1/factor #replace the zero cell
    
      
    comp = poisson.rvs(total_tst) 


    pmf = poisson.pmf(total_trn, total_tst)
    pmf_comp = poisson.pmf(comp, total_tst)

    cdf = poisson.cdf(total_trn, total_tst)
    cdf_comp = poisson.cdf(comp, total_tst)
    
    sig = (total_trn - total_tst)/np.sqrt(total_tst)
    sig_comp = (comp - total_tst)/np.sqrt(total_tst)
    
    if withsig:
        total_sig = np.ravel(np.sum(trn, axis = -1))
        sigcells = np.ravel_multi_index(np.where(trn[..., sigcol]!= 0), (10, 10, 10))
        print 'signal cells in training', total_sig[sigcells]
        print 'signal cells in testing', total_tst[sigcells]

        sig_sig = ((total_sig - total_tst)/np.sqrt(total_tst))[sigcells]
        cdf_sig = poisson.cdf(total_sig[sigcells], total_tst[sigcells])
        pmf_sig = poisson.pmf(total_sig[sigcells], total_tst[sigcells])
        

    bins, edges, patches = plt.hist(cdf, bins = 100, alpha = 0.6, color = 'orange', normed = False, histtype = 'stepfilled')
    plt.hist(cdf_comp, bins = edges, alpha = 0.4, color = 'blue', normed = False, histtype = 'stepfilled')
    if withsig:
        plt.hist(cdf_sig, bins = edges, alpha = 0.6, color = 'red', normed = False)
    plt.title('cumulative distributive function')
    plt.ylabel('number of cells')
    plt.xlabel('cdf')
    plt.ylim([1, 200])
    plt.yscale('log')
    plt.show()
    
    bins, edges, patches = plt.hist(pmf, bins = 100, alpha = 0.5, color = 'orange', histtype = 'stepfilled')
    plt.hist(pmf_comp, bins = edges, alpha = 0.4, color = 'blue', histtype = 'stepfilled')
    if withsig:
        plt.hist(pmf_sig, bins = 100, alpha = 0.7, color = 'red', histtype = 'stepfilled')
    plt.title('probability mass function')
    plt.xlabel('probability mass function')
    plt.ylabel('number of cells')
    plt.yscale('log')
    plt.ylim([0.8, 1000])
    plt.show()

    bins, edges, patches = plt.hist(np.log(pmf), bins = 100, alpha = 0.5, color = 'orange', histtype = 'stepfilled')
    plt.hist(np.log(pmf_comp), bins = edges, alpha = 0.4, color = 'blue', histtype = 'stepfilled')
    if withsig:
        plt.hist(np.log(pmf_sig), bins = 100, alpha = 0.8, color = 'red', histtype = 'stepfilled')
    plt.title('log probability mass function')
    plt.xlabel('log probability mass function')
    plt.ylabel('number of cells')
    plt.yscale('log')
    plt.ylim([0.7, 1000])
    plt.show()
    
    bins, edges, patches = plt.hist(np.ravel(sig), bins = 100, alpha = 0.4, color = 'orange', histtype = 'stepfilled')
    plt.hist(sig_comp, bins = edges, alpha = 0.4, color = 'blue', histtype = 'stepfilled')
    if withsig:
        plt.hist(sig_sig, bins = 100, alpha = 0.8, color = 'red', histtype = 'stepfilled')
    plt.title('deviation = (trn - tst)/sqrt(tst)')
    plt.xlabel('deviation')
    plt.ylabel('number of cells')
    plt.yscale('log')
    plt.ylim([0.8, 100])
    plt.show()
        
# A method that returns a function of mu, un-normalised
def negloglikelihood(bkg, sig, bkg_weights, sig_weight):
    '''
    bkg: number of each type of background events in each cell
    sig: number of signal events in each cell 
    bkg weights: the ratios we need to use to adjust them to realistic ratios 
    sig weight: the fraction we need to multiply the signal by to make it a function of mu
    '''
    
    # Add weights to the arrays 
    assert bkg.shape[-1] == len(bkg_weights)
    
    b = np.sum(bkg * np.array(bkg_weights), axis = -1)
    s = np.sum(sig * np.array(sig_weight), axis = -1)
    print 'background:', np.sum(b) 
    print 'signal:', np.sum(s) 
    
    assert b.shape == s.shape 
    print 'number of zero cells:', np.sum(b == 0) 
    print 'zero cells in sig:', s[b==0]
    b[b==0] = np.max(bkg_weights)
    print 'zero cells are replaced with', np.max(bkg_weights)
    
    print 'signal cells in bkg:', b[s!= 0] 
    print 'signal cells in sig:', s[s!= 0]
    
    return (lambda mu: 2 * np.sum((b + mu * s) - b * np.log(b + mu * s))) 


def negloglikelihood_plot(mu_range, fun, n = 100):
    ''' 
    a function that takes the outcome of the likelihood function and plot the likelihood within a certain range of mu
    mu_range: a tuple
    fun: a function
    ''' 
    mu_i, mu_f = mu_range
    mu = np.linspace(mu_i, mu_f, n)
    y = map(fun, mu) - fun(0)
    ref = np.full(y.shape, 2 * 1.92)
    
    plt.plot(mu, y)
    plt.plot(mu, ref)
    plt.title('negative likelihood plot')
    plt.xlabel('mu')
    plt.ylabel('- 2 * log(likelihood)')
    plt.show() 
    
    return y

def bar3D(bins): 
    dim = bins.shape[0]
    types = bins.shape[-1]
    x, y, z = np.indices((dim, dim, dim))
    for t in range(types):     
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.scatter(x, y, z, c = bins[..., t], norm = LogNorm(), lw = 0)
        plt.show()
```
