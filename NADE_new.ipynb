```python
# A method that deletes certain cols from data and feature list, useful for removing muons
def remove_cols(data, indices, features):
    features = np.delete(features, indices)
    n_features = len(features)
    data = np.delete(data, indices, 1) 
    return data, features 

# A method that replaces all the zeros and put them at an unrealistic value
def replace_zeros(dataset, mus = None, sigmas = None): 
    samples = dataset.shape[0] 
    features = dataset.shape[1] 
    maxs = dataset.max(axis = 0)
    mmax = maxs.max()
    copy = np.copy(dataset)
    copy[copy == 0] = mmax + 100
    mins = copy.min(axis = 0) 
    print maxs
    print mins
    if mus!= None and sigmas!= None:
        means = mus
        stds = sigmas
    else:
        stds = (maxs - mins)/5
        means = 2 * mins - maxs
    
    print stds

    for i in range(features): 
        for j in range(samples):
            if copy[j, i] == mmax + 100:
                copy[j, i] = np.random.normal(means[i], stds[i])
    return (mins, means, stds, copy)

# A method that splits the data into training and testing, and save the arrays to files 
def split(data, label, ratio = 0.5, io = True, show = False): 
    
    # split into right ratios 
    np.random.shuffle(data)
    dim = len(data)
    trndata = data[0:ratio * dim, :] 
    cvdata, tstdata = np.array_split(data[ratio * dim: dim-1, :], 2, axis = 0)
    
    # Replace the zeros 
    mins, mus, sigmas, trndata = replace_zeros(trndata)
    a, b, c, tstdata = replace_zeros(tstdata, mus, sigmas)
    a, b, c, cvdata = replace_zeros(cvdata, mus, sigmas)
    
    io_training = None 
    # Dump into hdf5 files
    if io:
        print label
        io_training = '../Data/%s_training.hdf5'%label 
        print io_training
        f = h5py.File(io_training, 'a')
        print f.keys()
        if "train" in f.keys():
            print 'train exists'
            grp = f['train']
            del grp['data']
            dset = grp.create_dataset('data', data= trndata)
            grp.attrs['mus'] = mus
            grp.attrs['mins'] = mins
            grp.attrs['sigmas'] = sigmas 
            del f['test']['data']
            del f['validation']['data']
            f.create_dataset('test/data', data = tstdata)
            f.create_dataset('validation/data', data = cvdata)
        else:
            print 'creating train'
            grp = f.create_group('train')
            grp.create_dataset('data', data = trndata)
            grp.attrs["mus"] = mus
            grp.attrs["mins"] = mins
            grp.attrs["sigmas"] = sigmas 
            f.create_group('validation').create_dataset('data', data = cvdata)
            f.create_group('test').create_dataset('data', data = tstdata) # don't waste too much data on testing 
        f.close() 
    
    # Check if everything makes sense 
    if show: 
        for i in range(data.shape[-1]):
            plt.hist(trndata[:, i], bins = 30, alpha = 0.4, normed = True)
            plt.hist(tstdata[:, i], bins = 30, alpha = 0.4, normed = True)
            plt.hist(cvdata[:, i], bins = 30, alpha = 0.4, normed = True)
            plt.show() 
    
    return trndata, tstdata, io_training

# A method that randomly takes a fracture of the data
def take(data, ratio): 
    np.random.shuffle(data)
    dim = len(data)
    data = data[0:ratio * dim, :] 
    return data 

# Create nade data from hdf5 files
def create(src):
    if not os.path.isfile(src):
        print 'not a file'
        return (None,0,0)
    
    f=h5py.File(src)
    p5=f.get("final_model").get("parameters")
    means=[]
    rmss=[]
    if f.get('normalisation/mean'):
        means = f.get('normalisation/mean').value
    if f.get('normalisation/std'):
        rmss = f.get('normalisation/std').value
    p={}
    for k in p5.keys():
        p[k] = p5.get(k).value

    clsn = p.get("__class__")
    cls = getattr(NADE,clsn)
    nade = cls.create_from_params( p )
    return (nade,means,rmss)

# Get a model from a oNADE object
def getModel(nade):
    directory = '%s/%s'%(nade.args['resultspath'], nade.args['label'])
    files = [os.path.join(directory,d) for d in os.listdir(directory)]
    training = max(files, key=os.path.getmtime)
    print training
    files.remove(training)
    dout = max(files, key = os.path.getmtime)
    print dout
    return (create(dout), dout)
    
# Get the pseudo data and recover the zeros 
def extract(model, mins, nsampling = 1000): 
    nade, means, rmss = model
    pseudo = (nade.sample(nsampling))
    
    means = means.reshape(len(means), 1)
    rmss = rmss.reshape(len(rmss), 1)
    
    pseudo = pseudo * rmss + means 
    pseudo = recover_zeros(pseudo.T, mins)
    return pseudo 

# A method that takes a model, rescale the tstdata according to it and generate pseudo data with same dim as tstdata
def compare(model, data, density = True, features = None): 
    
    # Unpack the model tuple
    nade, means, rmss = model
    means = means.reshape(len(means), 1)
    rmss = rmss.reshape(len(rmss), 1)
    
    # Rescale data
    data = ((data.T - means)/rmss)
    pseudo = nade.sample(data.shape[1])
    
    # Print out loglikelihood
    estimate = nade.estimate_loglikelihood_for_dataset(Data.Dataset(data.T))
    print estimate.estimation
    print estimate.se
    
    # Plot log density
    if density:
        logdensity = nade.logdensity(data)
        print 'anomalies:', logdensity[logdensity < -50]
        hist, edges, batches = plt.hist(logdensity, label = 'data', range = (-50, 0), bins = 100, alpha = 0.4, color = 'blue', histtype = 'stepfilled')
        plt.hist(nade.logdensity(pseudo), label = 'pseudo', bins = edges, alpha = 0.4, color = 'red', histtype = 'stepfilled')
        plt.yscale('log')
        plt.ylim([0.5, 1000])
        plt.legend()
        plt.show()
        
    # Plot histgrams if features 
    if features != None:
        for i in range(len(features)):
            print 'feature:', features[i]
            hist, edges, patches = plt.hist(data[i, :], bins = 100, alpha = 0.5, histtype = 'stepfilled')
            plt.hist(pseudo_tst[i, :], bins = edges, alpha = 0.5, histtype = 'stepfilled')
            plt.show()

    return data, pseudo  
```
